```python
#Import necessary libraries
import numpy as np
import pandas as pd
```

```python
#import the datasets
crime  = pd.read_csv("crimebystatecombinedwithunemployment.csv")
states = pd.read_csv("states.csv")
```

In order to make much sense of the state abbreviatio, we will include a column to hold the full state 
name thereby clarifying the abbreviated forms. To do this we make the colums for the two data frames thesame to enable an quick index merge.

```python
#we set the index for the crime df to "state"
c = crime.set_index("state")
c.head()
```

```python
#Set the index of the states df to "Abbreviation"
s = states.set_index('Abbreviation')
s.head()
```

```python
#expand all rows
pd.set_option("display.max_rows", None)
```

```python
#we join and reset index to obtain our new df containing the full names of the 50 states in the US
crime = c.join(s)
crime = crime.reset_index()
crime = crime.rename({"index":"state_abbr"}, axis=1)
crime
```

```python
crime.to_csv("updated_crime.csv")
```
